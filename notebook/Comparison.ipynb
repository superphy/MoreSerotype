```python
import pandas as pd
import numpy as np
import json
%matplotlib inline
```

```python
# Read old ectyper dict
ectyper_dict = json.load(open('old_ectyper_dict.json'))
temp_list = []
for antigen, alleles in ectyper_dict.items():
    for name, allele in alleles.items():
        new_entry = {
            'serotype': allele.get('allele'),
            'name': name,
            "gene": allele.get('gene')
        }
        temp_list.append(new_entry)
old_ectyper_df = pd.DataFrame(temp_list)
```

```python
# Read ectyper dict
ectyper_dict = json.load(open('ectyper_dict.json'))
temp_list = []
for antigen, alleles in ectyper_dict.items():
    for name, allele in alleles.items():
        new_entry = {
            'serotype': allele.get('allele'),
            'name': name,
            'gene': allele.get('gene')
        }
        temp_list.append(new_entry)
ectyper_df = pd.DataFrame(temp_list)
```

```python
# Compare
merged_df = old_ectyper_df.merge(ectyper_df, left_on='name', right_on='name', how='outer', indicator=True)
diff_df = merged_df[merged_df['_merge']!='both']
```

```python
# Gene removed in latest iteration
diff_df[diff_df['gene_x'].notnull()]['gene_x'].value_counts().plot(kind='bar')
```

```python
# Gene added in latest iteration
diff_df[diff_df['gene_y'].notnull()]['gene_y'].value_counts().plot(kind='bar')
```
